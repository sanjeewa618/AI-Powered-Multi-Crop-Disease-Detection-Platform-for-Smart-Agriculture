# 02 - Dataset Visualization

**Goal:** Visually inspect PlantVillage dataset images — check image quality, clarity, disease visibility, and background consistency.

| Info | Value |
|------|-------|
| Dataset | PlantVillage |
| Task | Multi-Crop Disease Detection |
| Phase | 2 — Data Visualization |

```python
import os
import random
import matplotlib.pyplot as plt
from PIL import Image
```

```python
DATASET_PATH = "../dataset/PlantVillage"

classes = [
    folder for folder in os.listdir(DATASET_PATH)
    if os.path.isdir(os.path.join(DATASET_PATH, folder))
]

print("Total Classes:", len(classes))
```

```python
# Display a single random image
random_class = random.choice(classes)

class_path = os.path.join(DATASET_PATH, random_class)

images = os.listdir(class_path)

random_image = random.choice(images)

image_path = os.path.join(class_path, random_image)

image = Image.open(image_path)

plt.imshow(image)
plt.title(random_class)
plt.axis("off")
plt.show()
```

```python
# Display a 3x3 grid of random images from random classes
fig, axes = plt.subplots(3, 3, figsize=(12, 12))
fig.suptitle("PlantVillage - Random Sample Images", fontsize=16, fontweight="bold")

for ax in axes.flatten():
    rand_class = random.choice(classes)
    class_path = os.path.join(DATASET_PATH, rand_class)
    rand_img = random.choice(os.listdir(class_path))
    img = Image.open(os.path.join(class_path, rand_img))
    ax.imshow(img)
    ax.set_title(rand_class.replace("___", "\n"), fontsize=7)
    ax.axis("off")

plt.tight_layout()
plt.show()
```
