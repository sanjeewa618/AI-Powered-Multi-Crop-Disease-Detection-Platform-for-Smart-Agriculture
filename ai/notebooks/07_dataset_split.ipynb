# 07 - Dataset Split (Train / Validation / Test)

**Goal:** Split the PlantVillage dataset into 3 sets for model training.

| Split | Ratio | Purpose |
|-------|-------|---------|
| **Train** | 70% | Model learning |
| **Validation** | 15% | Monitor during training |
| **Test** | 15% | Final evaluation |

| Info | Value |
|------|-------|
| Dataset | PlantVillage |
| Task | Multi-Crop Disease Detection |
| Phase | 7 — Dataset Split |

```python
import splitfolders
import os

input_folder  = "../dataset/PlantVillage"
output_folder = "../dataset"

print("Input  :", input_folder)
print("Output :", output_folder)
print("Ratio  : Train 70% | Validation 15% | Test 15%")
print()

# Check input exists
if not os.path.exists(input_folder):
    print("❌ Dataset folder not found! Check the path.")
else:
    classes = [f for f in os.listdir(input_folder) if os.path.isdir(os.path.join(input_folder, f))]
    print(f"✅ Found {len(classes)} classes in PlantVillage dataset.")
    print("Ready to split...")
```

```python
# ⚠️ This cell copies ~54,000 images — may take 2–5 minutes
print("Starting dataset split...")
print("⏳ Please wait, this may take a few minutes...")
print()

splitfolders.ratio(
    input_folder,
    output=output_folder,
    seed=42,
    ratio=(0.70, 0.15, 0.15)
)

print()
print("✅ Dataset split complete!")
```

```python
# Verify split folders exist and count classes
print("=" * 50)
print("Split Verification")
print("=" * 50)

train_path = "../dataset/train"
val_path   = "../dataset/val"
test_path  = "../dataset/test"

print("Train Classes      :", len(os.listdir(train_path)))
print("Validation Classes :", len(os.listdir(val_path)))
print("Test Classes       :", len(os.listdir(test_path)))
```

```python
# Count images in each split
def count_images(folder):
    total = 0
    for cls in os.listdir(folder):
        cls_path = os.path.join(folder, cls)
        if os.path.isdir(cls_path):
            total += len([
                f for f in os.listdir(cls_path)
                if f.lower().endswith((".jpg", ".jpeg", ".png"))
            ])
    return total

train_count = count_images(train_path)
val_count   = count_images(val_path)
test_count  = count_images(test_path)
total       = train_count + val_count + test_count

print("=" * 50)
print("Image Count per Split")
print("=" * 50)
print(f"Train      : {train_count:>6} images  ({train_count/total*100:.1f}%)")
print(f"Validation : {val_count:>6} images  ({val_count/total*100:.1f}%)")
print(f"Test       : {test_count:>6} images  ({test_count/total*100:.1f}%)")
print("-" * 50)
print(f"Total      : {total:>6} images")
print("=" * 50)
print("\n✅ Dataset is ready for model training!")
```
