# 03 - Image Statistics

**Goal:** Analyze image dimensions, formats, and resolution consistency across the PlantVillage dataset before preprocessing.

| Info | Value |
|------|-------|
| Dataset | PlantVillage |
| Task | Multi-Crop Disease Detection |
| Phase | 3 — Image Statistics |

```python
import os
from PIL import Image
import pandas as pd

DATASET_PATH = "../dataset/PlantVillage"
```

```python
image_data = []

for class_name in os.listdir(DATASET_PATH):

    class_path = os.path.join(DATASET_PATH, class_name)

    if not os.path.isdir(class_path):
        continue

    for image_name in os.listdir(class_path):

        image_path = os.path.join(class_path, image_name)

        try:
            img = Image.open(image_path)

            width, height = img.size

            image_data.append({
                "Class": class_name,
                "Width": width,
                "Height": height,
                "Format": img.format
            })

        except:
            pass

df = pd.DataFrame(image_data)

df.head(10)
```

```python
print("Total Images :", len(df))

print("\nImage Formats")
print(df["Format"].value_counts())

print("\nAverage Width  :", round(df["Width"].mean(), 2))
print("Average Height :", round(df["Height"].mean(), 2))

print("\nMin Width  :", df["Width"].min(), "| Max Width  :", df["Width"].max())
print("Min Height :", df["Height"].min(), "| Max Height :", df["Height"].max())
```

```python
# Check how many unique sizes exist
df["Size"] = df["Width"].astype(str) + "x" + df["Height"].astype(str)

print("Unique Image Sizes:")
print(df["Size"].value_counts())
```
